```
from pyspark.sql.functions import format_number
```

```
data=stock.select(stock['Date'],
             format_number(stock['High'].cast('float'),2).alias('HIGH'),
             format_number(stock['Open'].cast('float'),2).alias('OPEN'),
             format_number(stock['Low'].cast('float'),2).alias('LOW'),
             format_number(stock['Close'].cast('float'),2).alias('CLOSE'),
             format_number(stock['Volume'].cast('float'),2).alias('VOLUME'),
             format_number(stock['Adj Close'].cast('float'),2).alias('Adj Close')
            )
```

```
df3=spark.read.csv('/FileStore/tables/sales_data.csv',header=True,inferSchema=True)
```

```
df3.show()
```

```
df3.printSchema()
```

```
df3=df3.withColumn('Total Cost',(df3['Quantity Ordered']*df3['Price Each']))
```

```
df3.show()
```

```
from pyspark.sql.functions import to_timestamp
```

```
df3=df3.withColumn('Order Date', to_timestamp(df3['Order Date'], 'MM/dd/yy HH:mm'))
```

```
df3.printSchema()
```

```
df3=df3.withColumn('Hour', hour(df3['Order Date']))
```

```
df3.show()
```

```
df3.printSchema()
```

```
from pyspark.sql.functions import sum
```

```
#1st Question
df3.groupBy('Month').sum('Total Cost').orderBy('sum(Total Cost)', ascending=False).show()
#12
```

```
#2nd Question
df3.groupBy('City').sum('Quantity Ordered').orderBy('sum(Quantity Ordered)', ascending=False).show()
#San Francisco
```

```
#3rd Question
df3.groupBy('Hour').sum('Total Cost').orderBy('sum(Total Cost)',ascending=False).show()
#7 p.m.
```

```
#4th Question
data=df3.groupBy('Order ID').count().where('count > 1').drop('count')
data1=list(data.select('Order ID').toPandas()['Order ID'])
df4=df3.filter(df3['Order ID'].isin(data1))
```

```
df4.groupBy('Order ID', 'Product').sum('Quantity Ordered').orderBy('Order ID').show()
```

```
#5th Question
df3.groupBy('Product').sum('Quantity Ordered').orderBy('sum(Quantity Ordered)',ascending=False).show()
#AAA batteries 4 pack
```

```
from pyspark.sql.functions import count, collect_list, col
df3.groupBy('Order ID').agg(count('Order ID'), collect_list('Product').alias('coll_prd')).where(col('count(Order ID)') > 1).groupBy('coll_prd').count().orderBy('count', ascending=False).show(1, False)
```
